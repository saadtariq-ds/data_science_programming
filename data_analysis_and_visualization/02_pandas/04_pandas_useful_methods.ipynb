## Apply Methods

```python
df.info()
```

**Selecting last 4 digits of credit card number (CC Number)**

```python
def last_four(number):
  return str(number)[-4:]
```

```python
df["last_four"] = df["CC Number"].apply(last_four)
```

```python
df.head()
```

```python
df.info()
```

**Returning $ sign based on value of total bill column**

```python
df['total_bill']
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
  if price < 10:
    return "$"
  elif price >= 10 and price < 30:
    return "$$"
  else:
    return "$$$"
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

```python
df.head()
```

## Apply on Multiple Columns

### Lambda Expression

```python
def simple(num):
  return num * 2
```

```python
lambda num: num * 2
```

```python
df['total_bill'].apply(simple)
```

```python
df['total_bill'].apply(lambda num: num * 2)
```

```python
df.head()
```

```python
def quality(total_bill, tip):
  if tip / total_bill > 0.25:
    return "Generous"
  else:
    return "Other"
```

```python
df["quality"] = df[["total_bill", "tip"]].apply(lambda df: quality(df["total_bill"], df["tip"]), axis=1)
```

```python
df.head()
```

### Vectorizer Method

```python
df['quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```

```python
df.head()
```
