# DataFrame

A DataFrame is a table of columns and rows in pandas that we can easily restructure and filter.

**Formal Definition:** A group of Pandas Series object that share the same index

```python
import numpy as np
import pandas as pd
```

## Creating a DataFrame

### Creating DataFrame from NumPy

```python
np.random.seed(101)
my_data = np.random.randint(0, 101, (4,3))
my_data
```

```python
my_index = ["CA", "NY", "AZ", "TX"]
```

```python
my_columns = ["January", "February", "March"]
```

```python
df = pd.DataFrame(data=my_data)
df
```

```python
df = pd.DataFrame(data=my_data, index=my_index)
df
```

```python
df = pd.DataFrame(data=my_data, index=my_index, columns=my_columns)
df
```
