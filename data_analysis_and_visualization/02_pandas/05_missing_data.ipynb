# Missing Data

Options for dealing with Missing Data
1. Keep it
2. Remove it
3. Replace it

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movie_scores.csv')
```

```python
df.head()
```

## Checking Null Values

```python
df.isnull() # returns True if a null value and false if not a null value
```

```python
df.notnull() # returns True if not a null value and false if a null value
```

```python
df['pre_movie_score'].notnull()
```

```python
df[df['pre_movie_score'].notnull()]
```

```python
df[(df['pre_movie_score'].isnull()) & (df['first_name'].notnull())]
```

## Keeping the Missing Data

```python
df.head()
```

## Removing the Missing Data

```python
df.head()
```

```python
df.dropna()
```

### Thresh argument

```python
df.dropna(thresh=1)
```

```python
df.dropna(thresh=4)
```

```python
df.dropna(thresh=5)
```

### Axis argument

```python
df.head()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=0)
```

### Subset Argument

```python
df.head()
```

```python
df.dropna()
```

```python
df.dropna(subset=['last_name'])
```

## Filling Missing Values

```python
df.head()
```

### Filling with Random Value

```python
df.fillna('NEW VALUE!')
```

### Filling with Zero

```python
df['pre_movie_score']
```

```python
df['pre_movie_score'].fillna(0)
```

### Filling with Mean

```python
df['pre_movie_score'].fillna(df['pre_movie_score'].mean())
```

```python
df.head()
```

```python
df.fillna(df.mean())
```

### Filling with Interpolate Method

```python
airline_ticket = {"first": 100, "business": np.nan, "economy-plus": 50, "economy": 30}
```

```python
series = pd.Series(airline_ticket)
```

```python
series
```

```python
series.interpolate()
```
