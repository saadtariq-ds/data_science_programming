# Distribution Plots

Distribution plots displays a single continuous feature and help visualize properties such as deviation and average values.

There are 3 main distribution plot types:
1. Rug Plot
2. Histogram
3. KDE Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df
```

## Rug Plot

A rug plot is the simpliest distribution plot and merely adds a dash or tick line for every single value.

The y-axis does not really have a meaning

```python
sns.rugplot(x='salary', data=df)
```

```python
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
plt.figure(figsize=(5,3), dpi=150)
sns.rugplot(x='salary', data=df, height=0.5)
```

## Distribution Plot

```python
sns.displot(data=df, x='salary')
```

```python
sns.displot(data=df, x='salary', bins=10)
```

```python
sns.displot(data=df, x='salary', bins=100)
```

### Changing Background Style

```python
sns.set(style='darkgrid')
sns.displot(data=df, x='salary', bins=10)
```

```python
sns.set(style='whitegrid')
sns.displot(data=df, x='salary', bins=10)
```

```python
sns.set(style='dark')
sns.displot(data=df, x='salary', bins=10)
```

```python
sns.set(style='white')
sns.displot(data=df, x='salary', bins=10)
```

```python
sns.set(style='ticks')
sns.displot(data=df, x='salary', bins=10)
```

### Different Parameters

```python
sns.displot(data=df, x='salary', color='red')
```

```python
sns.displot(data=df, x='salary', color='red', edgecolor='blue', linewidth=1)
```

### Combining Distplot, KDE plot and Rug plot

```python
sns.displot(data=df, x='salary', kde=True, rug=True)
```

## Histogram

```python
sns.histplot(data=df, x='salary')
```

## KDE Plot

Seaborn also allows us to add on a KDE plot curve on top of a histogram

KDE stands for Kernel Density Estimation. It is a method of estimating a probability density function of a random variable

In simpler terms, it is a way of estimating a continuous probability curve for a finite data sample

```python
sns.kdeplot(data=df, x='salary')
```

### Testing KDE plot

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages)
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30)
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x='age')
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100])
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.01)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.1)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.1, fill=True)
```
