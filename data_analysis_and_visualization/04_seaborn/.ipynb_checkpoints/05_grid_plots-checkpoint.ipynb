# Grid Plots

Seaborn grid call use Matplotlib subplots() to automatically create a grid based off a categorical column. Instead of passing in a specific number of rows and column for the subplots, we can simply pass in the name of the column and seaborn will automatically map the subplots grid

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

## Catplot

```python
sns.catplot(data=df, x='gender', y='math score', kind='box')
```

### Different Parameters

**Row Parameter**

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', row='lunch')
```

**Column Parameter**

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch')
```

**Using Both Row and Column**

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
```

**Playing with Kind Parameter**

```python
sns.catplot(data=df, x='gender', y='math score', kind='violin', col='lunch', row='test preparation course')
```

## Pairgrid

```python
sns.pairplot(data=df)
```

```python
sns.PairGrid(data=df)
```

```python
g = sns.PairGrid(data=df)
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot)
```

### Different Parameters

**Hue Parameter**

```python
g = sns.PairGrid(data=df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.kdeplot)
```

### Adding a Legend

```python
g = sns.PairGrid(data=df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.kdeplot)
g = g.add_legend()
```
