# Matrix Plots

Matrix plots are the visual equivalent of displaying a pivot table

The matrix plot displays all the data passed in, visualizing all the numeric values in a DataFrame

The two main types of matrix plots are:
1. heatmap
2. clustermap

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df
```

## Heatmap

Heatmap visually displays the distribution of cell values with a color mapping

```python
df = df.set_index("Countries")
df
```

```python
sns.heatmap(data=df)
```

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1))
```

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidths=0.5)
```

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True)
```

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, cmap='viridis')
```

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, cmap='viridis', center=40)
```

## Clustermap

Same visualization as heatmap, but first conduct hierarchical clustering to reorganize data into groups

```python
sns.clustermap(data=df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, cmap='viridis')
```

```python
sns.clustermap(data=df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, cmap='viridis', col_cluster=False)
```
