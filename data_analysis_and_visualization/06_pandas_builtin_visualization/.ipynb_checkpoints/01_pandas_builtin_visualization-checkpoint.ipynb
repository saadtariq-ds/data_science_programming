# Pandas Builtin Visualization

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

## Histograms

```python
df1['A'].hist()
```

```python
df1['A'].hist(bins=25)
```

## Plotting

### Histogram

```python
df1['A'].plot(kind='hist', bins=25)
```

```python
df1['A'].plot.hist()
```

### Area Plot

```python
df2.plot.area()
```

```python
df2.plot.area(alpha=0.4)
```

### Bar Plot

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

### Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

### Box Plots

```python
df2.plot.box()
```

### Hexagonal Bin Plot

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

### KDE Plot

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
