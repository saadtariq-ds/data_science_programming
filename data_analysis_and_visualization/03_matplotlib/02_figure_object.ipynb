# Figure Object

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Implementing Figures and Axes

```python
a = np.linspace(0, 10, 11)
```

```python
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

### Large and Small Axes

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

# Small Axes
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(x, y)
```

### Adding Specific Labels and Axes

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of 4")

# Small Axes
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(x, y)
axes2.set_xlim(0,2)
axes2.set_ylim(0,10)
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("Power of 2")
```

### Figure Parameters

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig = plt.figure(dpi=100)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig = plt.figure(figsize=(10, 3), dpi=100)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

### Exporting a Figure

```python
fig = plt.figure(figsize=(10, 3), dpi=100)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('object_figure.png')
```

```python
fig = plt.figure(figsize=(1, 3), dpi=100)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('object_figure2.png', bbox_inches='tight')
```
