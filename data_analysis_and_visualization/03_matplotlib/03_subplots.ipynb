# Subplots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = x * 2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].plot(x, y)
```

## Fixing Spaces between subplots

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].plot(x, y)

plt.tight_layout()
```

**Manually fixing spaces**

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)

fig.subplots_adjust(wspace=0.6, hspace=0.3)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][0].set_ylabel("Y Label 0,0")

axes[0][1].plot(a, b)
axes[0][1].set_ylabel("Y Label 0,1")

axes[1][0].plot(x, y)
axes[1][0].set_ylabel("Y Label 1,0")

axes[1][1].plot(a, b)
axes[1][1].set_ylabel("Y Label 1,1")

fig.suptitle('Figure Level', fontsize=10)

fig.subplots_adjust(wspace=0.6, hspace=0.3)
```

## Exporting Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][0].set_ylabel("Y Label 0,0")

axes[0][1].plot(a, b)
axes[0][1].set_ylabel("Y Label 0,1")

axes[1][0].plot(x, y)
axes[1][0].set_ylabel("Y Label 1,0")

axes[1][1].plot(a, b)
axes[1][1].set_ylabel("Y Label 1,1")

fig.suptitle('Figure Level', fontsize=10)

fig.subplots_adjust(wspace=0.6, hspace=0.3)

fig.savefig("subplots.png", bbox_inches='tight')
```
