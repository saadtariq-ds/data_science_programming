# Matplotlib Basics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

## Figuring Out Parameters

### X Label, Y Label and Title

```python
plt.plot(x, y)
plt.title("XYZ Title")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

### X Limits and Y Limits

```python
plt.plot(x, y)
plt.title("XYZ Title")
plt.xlabel("X axis")
plt.xlim(0, 6)
plt.ylabel("Y axis")
plt.ylim(0, 15)
```

## Saving a Plot

```python
plt.plot(x, y)
plt.title("XYZ Title")
plt.xlabel("X axis")
plt.xlim(0, 6)
plt.ylabel("Y axis")
plt.ylim(0, 15)
plt.savefig("matplotlib_basic_plot.png")
```
