# Styling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Legend

```python
x = np.linspace(0, 10, 11)
```

```python
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x)
ax.plot(x,y)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label="X vs X")
ax.plot(x, y, label="X vs Y")
ax.legend()
```

### Changing Position of Legend

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label="X vs X")
ax.plot(x, y, label="X vs Y")
ax.legend(loc='lower left')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label="X vs X")
ax.plot(x, y, label="X vs Y")
ax.legend(loc=4)
```

### Manually Setting Location

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label="X vs X")
ax.plot(x, y, label="X vs Y")
ax.legend(loc=(1.1, 0.5))
```

## Colors and Styles

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x, x)
```

### Changing Colors

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, color='purple')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, color='green')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, color='#650dbd')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, color='#e6c017')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, color='#e6c017', label='X vs X')
axes.plot(x, y, color='#545255', label='X vs Y')
axes.legend()
```

### Line

**Line Width**

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, color='#e6c017', lw=10)
axes.plot(x, y, color='#545255', linewidth=0.5)
axes.legend()
```

**Line Style**

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# possible linestyle options '--', '-', '-.', ':', 'steps'
axes.plot(x, x, color='#e6c017', lw=2, ls='--')
axes.plot(x, y, color='#545255', linewidth=2, linestyle='-.')
axes.legend() 
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color='#e6c017', lw=2, ls=':')
axes.plot(x, y, color='#545255', linewidth=2, linestyle='-.')
axes.legend() 
```

**Custom Linestyle**

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

lines = axes.plot(x, y, color='#545255', linewidth=3)
lines[0].set_dashes([5,2,10,2])
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

lines = axes.plot(x, y, color='#545255', linewidth=3)
lines[0].set_dashes([5,2,3,2,1,1])
```

### Markers

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, y, color='#545255', linewidth=3, marker='o')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, y, color='#545255', linewidth=3, marker='o', ls='--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, y, color='#545255', linewidth=3, marker='x', ls='--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color='#545255', linewidth=3, marker='x', ls='--')
axes.plot(x, y, color='#cfd143', linewidth=3, marker='+', ls='--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color='#545255', linewidth=3, marker='x', ls='--', markersize=10)
axes.plot(x, y, color='#cfd143', linewidth=3, marker='+', ls='--', ms=10)
```

**Marker Face Color, Edge Width and Edge Color**

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color='#545255', linewidth=3, marker='o', ls='--', markersize=10,
         markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color='#545255', linewidth=3, marker='s', ls='--', markersize=15,
         markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
```
