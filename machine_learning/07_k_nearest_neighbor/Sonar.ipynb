# K Nearest Neighbor - Sonar Dataset

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, \
                            recall_score, f1_score, ConfusionMatrixDisplay, PrecisionRecallDisplay, RocCurveDisplay
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.pipeline import Pipeline
```

## Import Dataset

```python
df = pd.read_csv('sonar.all-data.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.head()
```

## Exploratory Data Analysis

```python
sns.countplot(data=df, x='Label')
```

```python
plt.figure(figsize=(12,7), dpi=200)
sns.heatmap(df.corr(numeric_only=True), cmap='viridis')
```

```python
df['Target'] = df['Label'].map({'R':0,'M':1})
np.abs(df.corr(numeric_only=True)['Target']).sort_values().tail(6)
```

## Separating X and Y Variables

```python
X = df.drop(['Target','Label'],axis=1)
y = df['Label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Creating Pipeline

```python
scaler = StandardScaler()
```

```python
model = KNeighborsClassifier()
```

```python
operations = [("scaler", scaler), ("model", model)]
```

```python
model_pipeline = Pipeline(operations)
```

### Grid Model

```python
model.get_params().keys()
```

```python
k_values = list(range(1,30))
```

```python
param_grid = {'model__n_neighbors':k_values}
```

```python
grid_model = GridSearchCV(model_pipeline, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_.get_params()
```

### Accuracy against different K Values

```python
scores = grid_model.cv_results_['mean_test_score']
plt.plot(k_values, scores, 'o-')
plt.xlabel("K")
plt.ylabel("Accuracy")
```

### Prediction and Evaluation

```python
predictions = grid_model.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
matrix = confusion_matrix(y_test, predictions)
confusion_matrix_plot = ConfusionMatrixDisplay(confusion_matrix=matrix)
confusion_matrix_plot.plot()
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, predictions)}")
print(f"Precision Score: {precision_score(y_test, predictions, pos_label='M')}")
print(f"Recall Score: {recall_score(y_test, predictions, pos_label='M')}")
print(f"F1 Score: {f1_score(y_test, predictions, pos_label='M')}")
```
