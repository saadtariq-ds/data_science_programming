## Converting Categorical Columns to Dummy Variables

```python
df.info()
```

```python
df = df.drop('target', axis=1)
```

```python
df['type'] = pd.get_dummies(df['type'], drop_first=True)
```

```python
df.head()
```

## Separating X and Y Variables

```python
X = df.drop('quality', axis=1)
y = df['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

## Creating Pipeline

```python
model = SVC(class_weight='balanced')
```

```python
scaler = StandardScaler()
```

```python
operations = [("scaler", scaler), ("model", model)]
```

```python
model_pipeline = Pipeline(operations)
```

## Creating Grid Model

```python
C = [0.001, 0.01, 0.1, 0.05, 0.5, 1]
gamma = ['scale', 'auto']
kernel = ['rbf', 'linear']
```

```python
param_grid = {'model__C':C,
              'model__gamma':gamma,
              'model__kernel':kernel
             }
```

```python
grid_model = GridSearchCV(model_pipeline, param_grid=param_grid, scoring='accuracy')
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_params_
```

## Prediction and Evaluation

```python
predictions = grid_model.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
matrix = confusion_matrix(y_test, predictions)
confusion_matrix_plot = ConfusionMatrixDisplay(confusion_matrix=matrix)
confusion_matrix_plot.plot()
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, predictions)}")
print(f"Precision Score: {precision_score(y_test, predictions, pos_label='Fraud')}")
print(f"Recall Score: {recall_score(y_test, predictions, pos_label='Fraud')}")
print(f"F1 Score: {f1_score(y_test, predictions, pos_label='Fraud')}")
```

## Different Class Imbalance Techniques

```python
df['quality'].value_counts()
```

```python
class_count_0, class_count_1 = df['quality'].value_counts()

# Separate class
class_0 = df[df['quality'] == 'Legit']
class_1 = df[df['quality'] == 'Fraud']# print the shape of the class
print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```

### Function to Create a Model

```python
def balance_model(df):
    X = df.drop('quality', axis=1)
    y = df['quality']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
    
    model = SVC()
    scaler = StandardScaler()
    scaled_X_train = scaler.fit_transform(X_train)
    scaled_X_test = scaler.transform(X_test)
    
    C = [0.001, 0.01, 0.1, 0.05, 0.5, 1]
    gamma = ['scale', 'auto']
    kernel = ['rbf', 'linear']
    
    param_grid = {'C':C,
                  'gamma':gamma,
                  'kernel':kernel
                 }
    
    grid_model = GridSearchCV(model, param_grid=param_grid, scoring='accuracy')
    grid_model.fit(scaled_X_train, y_train)
    predictions = grid_model.predict(scaled_X_test)
    
    print(f"Accuracy Score: {accuracy_score(y_test, predictions)}")
    print(f"Precision Score: {precision_score(y_test, predictions, pos_label='Fraud')}")
    print(f"Recall Score: {recall_score(y_test, predictions, pos_label='Fraud')}")
    print(f"F1 Score: {f1_score(y_test, predictions, pos_label='Fraud')}")
```

### Method 1: Random Undersampling

Undersampling can be defined as removing some observations of the majority class. This is done until the majority and minority class is balanced out.

Undersampling can be a good choice when you have a ton of data -think millions of rows. But a drawback to undersampling is that we are removing information that may be valuable.

```python
class_0_under = class_0.sample(class_count_1)
```

```python
undersampled_df = pd.concat([class_0_under, class_1], axis=0)
```

```python
df['quality'].value_counts()
```

```python
undersampled_df['quality'].value_counts()
```

```python
balance_model(undersampled_df)
```

### Method 2: Random Oversampling

Oversampling can be defined as adding more copies to the minority class. Oversampling can be a good choice when you dont have a ton of data to work with.

A con to consider when undersampling is that it can cause overfitting and poor generalization to your test set.

```python
class_1_over = class_1.sample(class_count_0, replace=True)
```

```python
oversampled_df = pd.concat([class_1_over, class_0], axis=0)
```

```python
df['quality'].value_counts()
```

```python
oversampled_df['quality'].value_counts()
```

```python
balance_model(oversampled_df)
```

### Method 3: Random Under-Sampling With Imblearn

RandomUnderSampler is a fast and easy way to balance the data by randomly selecting a subset of data for the targeted classes. Under-sample the majority class(es) by randomly picking samples with or without replacement.

```python
random_under_sampler = RandomUnderSampler(random_state=101)
```

```python
X_train_undersampled, y_train_undersampled = random_under_sampler.fit_resample(X_train, y_train)
```

```python
random_under_sampler_df = pd.concat([X_train_undersampled, y_train_undersampled], axis=1)
```

```python
df['quality'].value_counts()
```

```python
random_under_sampler_df['quality'].value_counts()
```

```python
balance_model(random_under_sampler_df)
```

## Method 4: Random Over-Sampling With imblearn

One way to fight imbalanced data is to generate new samples in the minority classes. The most naive strategy is to generate new samples by random sampling with the replacement of the currently available samples. The RandomOverSampler offers such a scheme.

```python
random_over_sampler = RandomOverSampler(random_state=101)
```

```python
X_train_oversampled, y_train_oversampled = random_over_sampler.fit_resample(X_train, y_train)
```

```python
random_over_sampler_df = pd.concat([X_train_oversampled, y_train_oversampled], axis=1)
```

```python
df['quality'].value_counts()
```

```python
random_over_sampler_df['quality'].value_counts()
```

```python
balance_model(random_over_sampler_df)
```

### Method 5: Synthetic Minority Oversampling Technique (SMOTE)

This technique generates synthetic data for the minority class.

SMOTE (Synthetic Minority Oversampling Technique) works by randomly picking a point from the minority class and computing the k-nearest neighbors for this point. The synthetic points are added between the chosen point and its neighbors.

```python
smote = SMOTE()
```

```python
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
smote_dataframe = pd.concat([X_train_smote, y_train_smote], axis=1)
```

```python
df['quality'].value_counts()
```

```python
smote_dataframe['quality'].value_counts()
```

```python
balance_model(smote_dataframe)
```
